```python
import pandas as pd
df = pd.read_csv('restaurants.csv')
df_meau = pd.read_csv('restaurant-menus.csv')
```

```python
df
```

```python
def get_num(row):
    arr = row['price'].split(' USD')
    return float(arr[0])

df_meau['price_numerical'] = df_meau.apply(get_num,  axis=1)
```

```python
df_meau_stats = df_meau.groupby('restaurant_id')['price_numerical'].median().reset_index()
```

```python
df = df.merge(df_meau_stats, left_on = 'id', right_on = 'restaurant_id')
```

```python
df = df.dropna()
```

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(df['price_numerical'], df['score'])
print(f"{slope} {intercept} {r_value} {p_value} {std_err}")
```

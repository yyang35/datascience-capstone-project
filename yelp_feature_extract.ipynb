```python
exploded_df = df.explode('categories_list')
category_postal_counts = exploded_df.groupby(['categories_list', 'postal_code']).size().reset_index(name='count')
```

```python
category_postal_map = dict(zip(zip(category_postal_counts['categories_list'], category_postal_counts['postal_code']), category_postal_counts['count']))

# Function to get counts for each category in a row
def get_counts(row):
    return [category_postal_map.get((cat, row['postal_code']), 0) for cat in row['categories_list']]

# Apply the function to each row
df['category_postal_counts'] = df.apply(get_counts, axis=1)
```

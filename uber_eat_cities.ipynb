```python
import pandas as pd
import numpy as np 


df = pd.read_csv('restaurants.csv')
```

```python
list = df['full_address']
```

```python
list
```

```python
some = []
for item in list:
    item = str(item)
    a = item.split(',')
    if len(a) >  3:
        some.append(a[-3])
    
```

```python
len(some)
```

```python
some
```

```python
import pandas as pd 

count_dict = {}

for location in some:
    if location in count_dict:
        count_dict[location] += 1
    else:
        count_dict[location] = 1
```

```python
sorted_count_dict = dict(sorted(count_dict.items(), key=lambda item: item[1], reverse=True))

for location, count in sorted_count_dict.items():
    print(f"{location}: {count}")
```
